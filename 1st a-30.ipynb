```python
mydict = {'b':2,'a':1,'c':3}

mylist = list(mydict.items())
print(mylist)

unsorted_list =[]

for x in mylist:
    unsorted_list.append(x[1])
print(unsorted_list)
unsorted_list.sort()
print(unsorted_list)

sorted_tuples = []
for x in unsorted_list:
    for y in mydict:
        if mydict[y] == x:
            sorted_tuples.append((y,mydict[y]))

print(sorted_tuples)
```

```python
d = {'a': 5, 'b': 2, 'c': 8, 'd': 1}

l = list(d.items())
print(l)
for i in range(len(l)):
    for j in range(i + 1, len(l)):
        if l[i][1] > l[j][1]:
            l[i], l[j] = l[j], l[i]

d = dict(l)

print(list(d.items()))
```
